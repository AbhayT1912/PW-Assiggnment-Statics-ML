1. What is hypothesis testing in statistics?

Ans:- Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves formulating two competing hypotheses—the null hypothesis (H₀) and the alternative hypothesis (H₁)—and using sample evidence to decide whether to reject the null hypothesis. The goal is to determine whether observed data is consistent with the assumption (null hypothesis) or suggests a statistically significant effect.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:- The null hypothesis (H₀) represents a statement of no effect or no difference. It assumes that any observed differences are due to chance.
The alternative hypothesis (H₁) suggests that there is a real effect or a significant difference.
In hypothesis testing, we assume H₀ is true and use data to determine whether to reject it in favor of H₁.

3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans:- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It sets the threshold for determining statistical significance—commonly set at 0.05 (5%).
If the p-value from the test is less than α, we reject H₀, suggesting the result is statistically significant.

4. What are Type I and Type II errors? Give examples of each.

Ans:- Type I Error: Rejecting a true null hypothesis (false positive).
Example: A medical test falsely indicates a person has a disease.

Type II Error: Failing to reject a false null hypothesis (false negative).
Example: A medical test fails to detect the disease when the person actually has it.

5. What is the difference between a Z-test and a T-test? Explain when to use each.

Ans:- Z-test is used when the population standard deviation is known and the sample size is large (n > 30).

T-test is used when the population standard deviation is unknown and/or the sample size is small (n ≤ 30).

Both tests compare sample statistics to population parameters to determine if there’s a significant difference.

```python
#6. Python program to generate a binomial distribution and plot a histogram

import numpy as np
import matplotlib.pyplot as plt

n, p = 10, 0.5
binom_data = np.random.binomial(n, p, 1000)

plt.hist(binom_data, bins=11, edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()
```

```python
#7. Hypothesis testing using Z-statistics in Python

import numpy as np
from scipy import stats

data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

z = (sample_mean - mu) / (sample_std / np.sqrt(n))
p = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-statistic:", z)
print("p-value:", p)
```

```python
#8. Simulate normal distribution and calculate 95% confidence interval

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=100)

mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

plt.hist(data, bins=20, edgecolor='black')
plt.title("Simulated Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(ci[0], color='red', linestyle='dashed', label='95% CI Lower')
plt.axvline(ci[1], color='green', linestyle='dashed', label='95% CI Upper')
plt.legend()
plt.show()
```

```python
#9.  Calculate Z-scores and visualize with histogram

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Sample data
data = np.random.normal(100, 15, 100)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram
plt.hist(z_scores, bins=20, edgecolor='black')
plt.title("Z-Score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
